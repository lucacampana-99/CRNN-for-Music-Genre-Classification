```python
#train_test_split of song split NB: two song splits belong to the same song must be in the same split(train or test (mutually exclusive)) 
X_train,X_t=[],[]
S_train,S_test=np.array([]),np.array([])
Y_train,Y_test=np.array([]),np.array([])

c=0
for i,y,z in zip(np.array(S),np.array(X),np.array(Y)):
  if(i[:-6] in list(X_test)):
    

    X_t.append(y)
    S_test=np.append(S_test,i)
    Y_test=np.append(Y_test,z)
  else:
    
    X_train.append(y)
    S_train=np.append(S_train,i)
    Y_train=np.append(Y_train,z)

  
```

```python
#classes encoding
label_encoder = LabelEncoder()
Y_train_enc = label_encoder.fit_transform(Y_train)
Y_test_enc=label_encoder.transform(Y_test)
```

```python
#min_max_scaler
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_train_scaled = min_max_scaler.fit_transform(X_train)

X_t_scaled = min_max_scaler.transform(X_t)
```

```python
#fit_predict_score
mySVC=SVC(C=8)
mySVC.fit(X_train_scaled,Y_train_enc)
pred=mySVC.predict(X_t_scaled)
f1_score(Y_test_enc, pred,average='macro')
```
